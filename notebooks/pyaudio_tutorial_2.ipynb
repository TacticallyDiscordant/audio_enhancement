```python
import pyaudio as pya
import struct
import time
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from tkinter import TclError

%matplotlib tk
```

```python
CHUNK = 1024 * 2  # amount of data processed at once
FORMAT = pya.paInt16
CHANNELS = 1  # mono
RATE = 44100
```

```python
fig, (ax, ax2 )= plt.subplots(2, figsize=(15, 8))

# pyaudio class instance
p = pya.PyAudio()

# stream object
stream = p.open(
    format=FORMAT,
    channels=CHANNELS,
    rate=RATE,
    input=True,
    output=True,
    frames_per_buffer=CHUNK
)

# variable for plotting
x = np.arange(0, 2 * CHUNK, 2)
x_fft = np.linspace(0, RATE, CHUNK)

# create a Line object with random data
line, = ax.plot(x, np.random.rand(CHUNK), '-', lw=2)
line_fft, = ax2.semilogx(x_fft, np.random.rand(CHUNK), '-', lw=2)
# axes formatting
ax.set_title('AUDIO WAVEFORM')
ax.set_xlabel('samples')
ax.set_ylabel('volume')
ax.set_ylim(-2048, 2048)
ax.set_xlim(0, 2 * CHUNK)
plt.setp(ax, xticks=[0, CHUNK, 2* CHUNK], yticks=[-1024, 0, 1024])


ax2.set_xlim(20, RATE / 2)

plt.show(block=False)

frame_count = 0
start_time = time.time()

while True:
    data = stream.read(CHUNK)
    data_int = struct.unpack(str(CHUNK) + 'h', data)
    line.set_ydata(data_int)
    
    y_fft = fft(data_int)
    line_fft.set_ydata(np.abs(y_fft[0:CHUNK]) * 2 / (256 * CHUNK))

    
    fig.canvas.draw()
    fig.canvas.flush_events()
    frame_count += 1
    
```
