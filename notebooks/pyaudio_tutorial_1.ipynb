```python
import pyaudio as pya
import struct
import numpy as np
import matplotlib.pyplot as plt

%matplotlib tk
```

```python
CHUNK = 1024 * 2  # amount of data processed at once
FORMAT = pya.paInt16
CHANNELS = 1  # mono
RATE = 44100
```

```python
p = pya.PyAudio()

stream = p.open(
    format=FORMAT,
    channels=CHANNELS,
    rate=RATE,
    input=True,
    output=True,
    frames_per_buffer=CHUNK
)

data = stream.read(CHUNK)
data_int = struct.unpack(str(CHUNK) + 'h', data)
```

```python
fig, ax = plt.subplots()

x = np.arange(0, 2*CHUNK, 2)
line, = ax.plot(x, np.random.rand(CHUNK))
ax.set_ylim(int(-1024*1.5), int(1024*1.5))
ax.set_xlim(0, CHUNK)
while True:
    data = stream.read(CHUNK)
    data_int = struct.unpack(str(CHUNK) + 'h', data)
    line.set_ydata(data_int)
    fig.canvas.draw()
    fig.canvas.flush_events()
```
